```python
rating_price_var = store[store.Rating >=2.5]
plt.figure(figsize=(18,5))
rating_price_var.Rating.value_counts().sort_values(axis=0,ascending = False).plot(kind="bar")
```

```python
plt.figure(figsize=(18,5))
rating_price_var['Android Ver'].value_counts().plot(kind='bar')
```

```python
rating_price_var = rating_price_var.rename(columns = {"Android Ver": "Android_Ver", 
                                  })
rating_price_var.columns
```

```python
plt.figure(figsize=(18,5))
rating_price_var.groupby(['Price']).Android_Ver.value_counts().nlargest(10).plot(kind='area',
                                        title="Rating per version of android",colormap='gist_rainbow')
plt.xlabel('Android Versions')
plt.ylabel('rating value counts')
plt.show()
```

```python
df = rating_price_var['Size']
df.head(6)
```

```python
plt.figure(figsize=(10,5))
df.replace(r'M','',regex=True,inplace = True)
df.value_counts().nlargest(10).plot(kind='barh',title="Size v/s Rating",colormap='jet')
plt.ylabel('size in MB')
plt.xlabel('rating value counts')
plt.show()
```
